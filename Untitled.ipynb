```python
import numpy as np
```

Nous commençons par écrire une fonction nous donnant la masse de la fusée au cours du temps.

```python
def mass(t):
    time=min(5,t)
    mass=150-20*time
    return mass
```

Nous écrivons la fonction nous donnant l'accélération au cours du temps:
$(m_s+m_p)\frac{dv}{dt}=\frac{dm_p}{dt}v_e-\frac{1}{2}\rho v^2 A C_D -(m_s+m_p)g$

```python
def acce(t,v):
    if t>5:
        BR=0
    else:
        BR=20
    M=mass(t)
    Ve=325
    rho=1.091
    A=np.pi*.5**2
    Cd=.15
    g=9.81
    acceleration=(Ve*BR-g*M-.5*rho*A*Cd*v*((v**2)**.5))/(M)
    return acceleration
    
```

```python
acce(34,34)
```

fonction donnant la vitesse au cours du temps

```python
def speed(a,v):
    speed = v+0.1*a
    return speed
```

```python
speed(34.10891891891892,0)
```

fonction donnant la hauteur au cours du temps

```python
def height(v,h):
    height=max(h+.1*v,0)
    return height
```

```python
height(10,.2)
```

nous avons maintenant nos trois fonction h,v,a. Il faut les insérer dans une boucle qui, par pas de .1 secondes, calculera leurs valeurs, et les stockera pour le calcul suivant, jusqu'a ce que la fusée soit retombée au sol

log sert a stocké les valeurs de a,v, et h à chaque pas.

```python
v=0
h=0.00000000001 #je considère que c'est une assez bonne approximation de 0 et ça facilite l'écriture de la condition de la boucle
t=0
log=[]
while h>0:
    a=acce(t,v)
    v=speed(a,v)
    h=height(v,h)
    pas=[h,v,a]
    t=t+.1
    log.append(pas)
print(log)
```

le code a l'air de fonctionner, je le copie en y apportant de légère modification pour répondre aux question et vérifier qu'il donne les bonne réponses.

```python
v=0
h=0.0000000000000001 #je considère que c'est une assez bonne approximation de 0 et ça facilite l'écriture de la condition de la boucle
t=0
log=[]
Vm=0
Tm=0
HVm=0
while h>0:
    a=acce(t,v)
    v=speed(a,v)
    h=height(v,h)
    pas=[h,v,a]
    t=t+.1
    log.append(pas)
    if Vm<v:
        Vm=v
        Tm=t
        HVm=h
    
print(Vm,Tm,HVm)
```
