```python
import numpy as np
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

Nous commençons par écrire une fonction nous donnant la masse de la fusée au cours du temps.

```python
def mass(t):#changer les valeurs en variables ?
    time=min(5.,t)
    mass=150.-20.*time
    return mass
```

Nous écrivons la fonction nous donnant l'accélération au cours du temps:
$(m_s+m_p)\frac{dv}{dt}=\frac{dm_p}{dt}v_e-\frac{1}{2}\rho v^2 A C_D -(m_s+m_p)g$

```python
def acce(t,v):
    if t>=5.:
        BR=0.
    else:
        BR=20.
    M=mass(t)
    Ve=325.
    rho=1.091
    A=np.pi*.5**2.
    Cd=.15
    g=9.81
    acceleration=(Ve*BR-g*M-.5*rho*A*Cd*v*((v**2)**.5))/(M) #norme : commande abs(v) --> bug ?
    return acceleration
    
```

```python
acce(34,34)
```

fonction donnant la vitesse au cours du temps

```python
def speed(a,v):
    speed = v+0.1*a
    return speed
```

```python
speed(34.10891891891892,0)
```

fonction donnant la hauteur au cours du temps

```python
def height(v,h):
    height=max(h+.1*v,0.)
    return height
```

```python
height(10,.2)
```

nous avons maintenant nos trois fonction h,v,a. Il faut les insérer dans une boucle qui, par pas de .1 secondes, calculera leurs valeurs, et les stockera pour le calcul suivant, jusqu'a ce que la fusée soit retombée au sol

log sert a stocker les valeurs de a,v, et h à chaque pas.

```python
v=0.
h=0. 
t=0.
log=[]
Tlog=[]
Hlog=[]
Slog=[]
Alog=[]
while h>0. or t==0:
    a=acce(t,v)
    v=speed(a,v)
    h=height(v,h)
    t=round(t+.1,2) #utilisationd de round pour éviter de se retrouver avec t=4.999999 et de faire une itération de trop avec du fuel
    pas=[t,h,v,a]
    
    Tlog.append(t)
    Hlog.append(h)
    Slog.append(v)
    Alog.append(a)
    log.append(pas)
for i in range(0,len(log)):
    print(Slog[i],Hlog[i],Tlog[i])
```

```python
# visualization of the path
pyplot.figure(figsize=(8,6))
pyplot.grid(True)
pyplot.xlabel(r't', fontsize=18)
pyplot.ylabel(r'h', fontsize=18)
pyplot.title('rocket trajectory, flight time = %.2f' % t, fontsize=18)
pyplot.plot(Tlog,Hlog, 'k--', lw=2);
```
